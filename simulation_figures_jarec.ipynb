```python
# Code for recreating simulation figures
df_1000 = pd.read_csv(path_to_df_1000)
df_500 = pd.read_csv(path_to_df_500)
df_250 = pd.read_csv(path_to_df_250)

# Assuming df1, df2, and df3 are your dataframes
x_values = df_1000["rotations"]  

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True)  # sharey=True for sharing the y-axis

# Replace 'boots_v' with the actual column name/parameter you want to plot
y_values_df1 = df_1000['boots_v']
y_values_df2 = df_500['boots_v']
y_values_df3 = df_250['boots_v']

# Plotting
line_df1, = axes[0].plot(x_values, y_values_df1, marker='o', linestyle='-', color='#576066', label='1000 (Equal)')  # Line color for df1 is 6b5ca5
line_df2, = axes[0].plot(x_values, y_values_df2, marker='o', linestyle='-', color='#f87666', label='500')  # Line color for df2 is 72195a
line_df3, = axes[0].plot(x_values, y_values_df3, marker='o', linestyle='-', color='#95a78d', label='250')  # Line color for df3 is 4c1036

# Add labels and title
axes[0].set_xticks(range(0, 361, 60))
axes[0].set_ylabel('Bootstrap Distance')
axes[0].set_title('Bootstrap')

# Plotting
axes[1].plot(x_values, df_1000['perms_v'], marker='o', linestyle='-', color='#576066', label='1000 (Equal)')  # Line color for df1 is 6b5ca5
axes[1].plot(x_values, df_500['perms_v'], marker='o', linestyle='-', color='#f87666', label='500')  # Line color for df2 is 72195a
axes[1].plot(x_values, df_250['perms_v'], marker='o', linestyle='-', color='#95a78d', label='250')  # Line color for df3 is 4c1036

# Add labels and title
axes[1].set_xticks(range(0, 361, 60))
axes[1].set_ylabel('Permutation Distance')
axes[1].set_title('Permutation')



axes[2].plot(x_values, df_1000['sss_v'], marker='o', linestyle='-', color='#576066', label='1000 (Equal)')  # Line color for df1 is 6b5ca5
axes[2].plot(x_values, df_500['sss_v'], marker='o', linestyle='-', color='#f87666', label='500')  # Line color for df2 is 72195a
axes[2].plot(x_values, df_250['sss_v'], marker='o', linestyle='-', color='#95a78d', label='250')  # Line color for df3 is 4c1036

# Add labels and title
axes[2].set_xticks(range(0, 361, 60))
axes[2].set_ylabel('SS Ratio')
axes[2].set_title('Sums of Squares Ratio')

fig.legend(handles=[line_df1, line_df2, line_df3], labels=['1000 (Equal)', '500', '250'],
           loc='upper center', bbox_to_anchor=(0.509, 1), ncol=3)

#Change title depending on what parameter you have
fig.text(0.509, 0.02, 'V Rotation', ha='center', va='center')
fig.suptitle('Effect of Rotating V', x=0.509, y=1.05, ha='center')

# Show the plot
plt.show()
fig.savefig('sims.pdf', dpi=300)
```

```python
# grid from varying c, r but holding V constant at 0
grd = pd.read_csv("0_boot_grid.csv",  index_col=0)
grd_0 = grd.apply(pd.to_numeric)

# grid from varying c, r but holding V constant at 1800
grd = pd.read_csv("180_boot_grid.csv",  index_col=0)
grd_180 = grd.apply(pd.to_numeric)

#set min/max colorations so all plots are on the same scale
vmin_boot = (grd_0.values.min())
vmax_boot = (grd_180.values.max())

for rotat in ["0", "45", "90", "180"]:
    grd = pd.read_csv(rotat+"_boot_grid.csv",  index_col=0) #set path to other parameter grid evaluations
    #grd = np.log(grd.apply(pd.to_numeric))
    cmap = sns.cm.crest_r
    s3 = sns.heatmap(grd, annot=True, fmt=".3g", vmin=vmin_boot, vmax=vmax_boot, annot_kws={"fontsize": 16}, cmap=cmap)
    s3.set_xlabel('Radius Shift', fontsize=12)
    s3.set_ylabel('Center Shift', fontsize=12)

    # Set the font size for tick labels on the x-axis
    s3.set_xticklabels(s3.get_xticklabels(), fontsize=12)

    # Set the font size for tick labels on the y-axis
    s3.set_yticklabels(s3.get_yticklabels(), fontsize=12)
    fig = s3.get_figure()
    fig.savefig("/Users/tarekzikry/Downloads/results_boot_grid_new"+'/boot_plot_'+rotat+'.pdf', dpi=300)
    plt.clf()
```
